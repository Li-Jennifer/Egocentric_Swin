```python
import json
import matplotlib.pyplot as plt

# 读取train.json文件
with open('swin_tiny_train.json', 'r') as f:
    swin_tiny_train = json.load(f)
    
with open('swin_small_train.json', 'r') as f:
    swin_small_train = json.load(f)  
    
with open('swin_base_train.json', 'r') as f:
    swin_base_train = json.load(f)      

# 提取epoch和loss数据
epochs = [item['epoch'] for item in swin_tiny_train]
swin_tiny_loss = [item['loss'] for item in swin_tiny_train]
swin_small_loss = [item['loss'] for item in swin_small_train]
swin_base_loss = [item['loss'] for item in swin_base_train]

# 绘制曲线图
plt.figure(figsize=(10, 6))
plt.plot(epochs, swin_tiny_loss, linestyle='-',label='swin_tiny_loss')
plt.plot(epochs, swin_small_loss, linestyle='-',label='swin_small_loss')
plt.plot(epochs, swin_base_loss, linestyle='-',label='swin_base_loss')

plt.title('Swin_Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# 读取val.json文件
with open('swin_tiny_val.json', 'r') as f:
    swin_tiny_val = json.load(f)
    
with open('swin_small_val.json', 'r') as f:
    swin_small_val = json.load(f)    
    
with open('swin_base_val.json', 'r') as f:
    swin_base_val = json.load(f)    
    
    
swin_tiny_val_top1_acc = [item["acc/top1"] for item in swin_tiny_val]
swin_tiny_val_top5_acc = [item["acc/top5"] for item in swin_tiny_val]

swin_small_val_top1_acc = [item["acc/top1"] for item in swin_small_val]
swin_small_val_top5_acc = [item["acc/top5"] for item in swin_small_val]

swin_base_val_top1_acc = [item["acc/top1"] for item in swin_base_val]
swin_base_val_top5_acc = [item["acc/top5"] for item in swin_base_val]

# 绘制曲线图
plt.figure(figsize=(10, 6))
plt.plot(epochs, swin_tiny_val_top1_acc, linestyle='-',label='swin_tiny_acc_top1')
plt.plot(epochs, swin_tiny_val_top5_acc, linestyle='-', label='swin_tiny_acc_top5')

plt.plot(epochs, swin_small_val_top1_acc, linestyle='-',label='swin_small_acc_top1')
plt.plot(epochs, swin_small_val_top5_acc, linestyle='-', label='swin_small_acc_top5')

plt.plot(epochs, swin_base_val_top1_acc, linestyle='-',label='swin_base_acc_top1')
plt.plot(epochs, swin_base_val_top5_acc, linestyle='-', label='swin_base_acc_top5')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Swin_Accuracy')
plt.legend()  # 显示图例
plt.show()
```

```python
swin_tiny_time = [item["time"] for item in swin_tiny_train]
swin_small_time = [item['time'] for item in swin_small_train]
swin_base_time = [item['time'] for item in swin_base_train]

# 绘制曲线图
plt.figure(figsize=(10, 6))
plt.plot(epochs, swin_tiny_time, linestyle='-',label='swin_tiny_time')
plt.plot(epochs, swin_small_time, linestyle='-',label='swin_small_time')
plt.plot(epochs, swin_base_time, linestyle='-', label='swin_base_time')

plt.xlabel('Epoch')
plt.ylabel('Time')
plt.title('Swin_Time')
plt.legend()  # 显示图例
plt.show()
```
